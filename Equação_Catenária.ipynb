```python
import numpy as np               # Import packages numpy ans matplotlib
import matplotlib.pyplot as plt
e=2,71828
a=1
```

```python
def f(x):    # Definition of function multisentence f
    return a*0.5*(np.exp(x/a)+np.exp(-x/a))
valor=f(1)
print (valor)
```

```python
x=np.linspace(-1,1)    # Comprehenssion list for values of x and y
y=[f(i) for i in x]
print(y)
```

```python
plt.plot(x,y)   #Ploting of graph of f
plt.grid()
plt.show()
```
