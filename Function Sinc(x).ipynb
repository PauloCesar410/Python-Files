```python
import numpy as np 
import math# Import packages numpy ans matplotlib
import matplotlib.pyplot as plt

def f(x):    # Definition of function multisentence f
    return (math.sin(x))/x
         

x=np.linspace(-15,15)    # Comprehenssion list for values of x and y
y=[f(i) for i in x]
#print(math.pi)
print(f(1))
plt.figure(figsize=(8, 8))
plt.title ('Função Seno Cardinal')
plt.plot(x,y, color='#8B0000', label='função sinc(x)=sin(x)/x')   #Ploting of graph of f
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np 
import math# Import packages numpy ans matplotlib
import matplotlib.pyplot as plt

def g(x):    # Definition of function multisentence f
    return ( math.sin(math.pi*(2*x-1)) - math.sin(math.pi*(x-0.5)) )/(math.pi*(x-0.5))
         

x1 = np.linspace(-6,6)    # Comprehenssion list for values of x and y
y1 = [g(i) for i in x1]
#print(y)
print(g(-1))
plt.figure(figsize=(8, 8))
plt.title ('Função wavelet sinclet')
plt.plot(x1,y1, color='#8B0000', label='Wavelet Function Sinclet ')   #Ploting of graph of f
plt.grid()
plt.legend()
plt.show()
```
