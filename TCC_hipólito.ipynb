```python
import numpy as np 
import matplotlib.pyplot as plt

# avalia o polinomio interpolador usando polinomios de Lagrange

def interpolL(x,T):

    S=0

    for i in range(len(T)):

        prod=1.0

        for j in range(len(T)):

            if j==i: continue

            prod=prod*(x-T[j][0])/(T[i][0]-T[j][0])

        S=S+T[i][1]*prod

    return S

T=[(0.0,0.0),(5.00,0.0),(5.20,0.0),(5.30, 0.025),(7.25,1.89),(8.55,2.856),(10.30,3.332),(10.55,3.667),(11.30,3.642),(11.50,0.94)]  # A tabela eh uma lista de pares ordenados.    

X=linspace(0,15,100)

Y= lambda X: interpolL(X,T) # define o polinomio interpolador p(x) 

plt.figure(figsize=(10,10))

plot(X,Y(x))

print(Y(7.00))

grid()

show()
```
