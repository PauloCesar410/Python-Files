```python
d=float(input("Digite o valor da distância em metros:  "))
```

```python
K=9*10**9
campo_eletrico=(q*K)/(d**2)
print(campo_eletrico,'N/m' )
```

# Gráfico de um Campo Elétrico

```python
import numpy as np
import matplotlib.pyplot as plt
def E1(d):
    return (K*q)/(d**2)  
```

```python
x1=np.linspace(1,10) 
y=[E1(d) for d in x1]
plt.plot(x1,y, color='#8B0000', label='função E(x)=(k*q)/d^2')
plt.grid()
plt.legend()
plt.show()
```

# Cálculo da derivada do potencial elétrico: campo elétrico na esfera.

```python
R=10
Pi=3.14159265
K=9*10**9
epsilon0=8.854*10**(-12)
sigma = 5*10**(-9)
Q=sigma*4*Pi*R**2
def potencial(r):
    return (K*Q)/r 
```

```python
dr=0.000001
def potencial_derivada(r):
    potencial_derivada = -(potencial(r-2*dr)+8*potencial(r+dr)-8*potencial(r-dr)-potencial(r+2*dr))/12*dr
    return potencial_derivada
```

```python
print(potencial_derivada(10))
```

# Gráfico do campo elétrico  na esfera usando derivadas 

```python
x2=np.linspace(1,10) 
y=[potencial_derivada(r) for r in x2]
plt.plot(x2,y, color='#9C0011', label='função E(x)=-dV/dr')
plt.grid()
plt.legend()
plt.show()
```

# Cálculo do campo elétrico nas proximidades da esfera

```python
R=float(input("Digite o valor do raio da esfera em metros:  "))
```

```python
dr=0.000001
r=R+dr
epsilon0=8.854*10**(-12)
sigma = 5*10**(-9)    #Densidade superficial de cargas C/m^2; sigma = Q/A
```

```python
def Campo(r):
    Campo = ((R/r)**2)*(sigma/epsilon0)
    return Campo
print(Campo(10))
```

# Gráfico do potencial elétrico na esfera usando a fórmula E(x)=(R/r)**2*(sigma/epsilon0)

```python
x=np.linspace(1,10) 
y=[Campo(r) for r in x]
plt.plot(x,y, color='#8B0000', label='função E(x)=(R/r)**2*(sigma/epsilon0)')
plt.grid()
plt.legend()
plt.show()
```

# Códigos para o artigo do CNMAC

```python
#Criando o gráfico da função potencial.
def Potencial(r):
    epsilon0=8.854*10**(-12)
    pi= 3.14159265314
    Q=0.000003
    if (0.01<r<=2):  #Aqui estou considerando uma esfera de raio r=2
         return (1/(4*pi*epsilon0))*(Q/2) 
    elif (r==2):
         return ((1/(4*pi*epsilon0))*(Q/r))
    else:
         return ((1/(4*pi*epsilon0))*(Q/r))
```

```python
print(Potencial(0.3))
print(Potencial(2))
print(Potencial(2.1))
```

```python
import numpy as np
import matplotlib.pyplot as plt 
x=np.linspace(1,10)
y=[Potencial(r) for r in x]
plt.plot(x,y,color='#A52A2A', label='função V(x)=K*Q/r', marker ="o")
plt.xlabel("Valor das distâncias r(metros)")
plt.ylabel("Valor do Potencial V(volts)")
plt.title("Gráfico do Potencial(V) versus a distância(r)")
plt.grid()
plt.legend()
plt.show()
```

```python
#Criando o gráfico da função campo elétrico usando derivada numérica.
def Campo_Eletrico(r):
    dr=0.01
    if (0<r<2):
        return 0
    else: 
         return  -(Potencial(r-2*dr)+8*Potencial(r+dr)-8*Potencial(r-dr)-Potencial(r+2*dr))/12*dr    
```

```python
print(Campo_Eletrico(1.98))
print(Campo_Eletrico(1.99))
print(Campo_Eletrico(2.00))
print(Campo_Eletrico(2.01))
```

```python
import numpy as np
import matplotlib.pyplot as plt 
x=np.linspace(1,10)
y=[ Campo_Eletrico(r) for r in x]
plt.plot(x,y,color='#32CD32', label='função V(x)=-dV/dr', marker ="o")
plt.xlabel("Valor das distâncias r(m)")
plt.ylabel("Valor do Campo Elétrico E(V/m)")
plt.title("Gráfico do Campo Elétrico(E) versus a distância(r)")
plt.grid()
plt.legend()
plt.show()
```
