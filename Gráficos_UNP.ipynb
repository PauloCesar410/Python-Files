```python
def H(thetac):
    return 0.3*thetac + 2
  
```

```python
import matplotlib.pyplot as plt
x=[0,5,10,15,20,25,30]    
y=[H(i) for i in x]
plt.plot(x,y,color='#8B0000', label='função Altura x Temperatura')
plt.ylabel("H(cm)", size = 16,)
plt.xlabel("Thetac(°C)", size = 16)
plt.grid()
plt.legend()
plt.show()
```

```python
def tensao(i):
    V = 1000*i
    return V
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,15)   
y=[tensao(i) for i in x]
plt.figure(figsize=(5, 5))
plt.title ('Gráfico da tensão verus corrente ')
plt.plot(x,y, color='#8B0000', label='V(i)=Ri') 
plt.xlabel("Valor da corrente elétrica(A)", size = 16,)
plt.ylabel("Valor da tensão elétrica(V)", size = 16)
plt.grid()
plt.legend()
plt.show()
```
