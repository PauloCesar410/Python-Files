```python
import numpy as np



def pend(y, t, p, to):
    arco, omega = y
    dydt    = [omega, (p/to)*(1+omega**2)**1/2]
    return dydt


p = 0.25
to = 5.0

yo = [0.01 - 0.1, 0.0]
t = np.linspace(0,10 ,101)

from scipy.integrate import odeint
aqua = odeint(pend, yo, t, args=(p,to))


from pandas import DataFrame
tempo = t
arco = aqua[:,0]
omega = aqua[:,1]
df = DataFrame({'a': arco, 'b': arco,'c': omega })
df.to_excel('---.xlsx',sheet_name='Sheet1', index=False)  
import matplotlib.pyplot as plt

plt.plot(t, aqua[:,0], 'b', label='arco(t)')
plt.plot(t, aqua[:,1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()


```
