```python
from pylab import *

# avalia o polinomio interpolador usando polinomios de Lagrange
def interpolL(x,T):
    S=0.0
    for i in range(len(T)):
        prod=1.0
        for j in range(len(T)):
            if j==i: continue
            prod=prod*(x-T[j][0])/(T[i][0]-T[j][0])
        S=S+T[i][1]*prod
    return S
T=[(0.38105, 234.605),(0.39973, 228.136),(0.40116, 228.103),(0.48742,234.509),(0.50389,239.768),(1.1496,253.52),(1.31328,257.674),(1.4023,259.384),(1.63359,265.435),] # A tabela eh uma lista de pares ordenados.
            
p= lambda x: interpolL(x,T) # define o polinomio interpolador p(x)

x=linspace(0,2,40)
#plt.ylim(0,280)
plot(x,p(x))
grid()
show()
#print(p(5))
```
