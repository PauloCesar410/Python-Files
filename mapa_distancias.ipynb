```python
#!pip install plotly
import math
import plotly.express as px

def leArquivo(nomeArquivo):
    arquivo = open(nomeArquivo, "r")
    linhas = arquivo.readlines()
    arquivo.close()
    return linhas


def imprimeArquivo(linhas):
    for l in linhas:
        l = l.rstrip("\n")
        print(l)


linhas = leArquivo("C:/Users/linha/2m6q.pdb")


# imprimeArquivo(linhas)

def obtemCadeia(cadeia, linhas):
    atoms = {}
    for l in linhas:
        l = l.rstrip("\n")
        if l[0:4] == "ATOM" and l[21] == cadeia and l[13:15] == "CA":
            resSeq = int(l[22:26])
            x = float(l[30:38])
            y = float(l[38:46])
            z = float(l[47:54])
            atoms[resSeq] = (x, y, z)
    return atoms


cadeia = obtemCadeia("A", linhas)
#print(cadeia)


def dist(p1, p2):
    x = 0
    y = 1
    z = 2
    return math.sqrt((p1[x] - p2[x]) ** 2 + (p1[y] - p2[y]) ** 2 + (p1[z] - p2[z]) ** 2)


def mapaDistancia(atomos):
    m = []
    i = 0
    for a1 in atomos:
        m.append([])
        for a2 in atomos:
            m[i].append(dist(atomos[a1], atomos[a2], ))
        i += 1
    return m

m = mapaDistancia(cadeia)
#print(m)

fig = px.imshow(m)
fig.show()
```
