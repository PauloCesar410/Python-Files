```python
import numpy as np               # Import packages numpy ans matplotlib
import matplotlib.pyplot as plt

def f(x):    # Definition of function multisentence f
    if (x>1):
        return 1
    elif (x<=1.0 and x>=-1.0):
        return x
    else:
        return -1 

x=np.linspace(-2,2)    # Comprehenssion list for values of x and y
y=[f(i) for i in x]
print(y)

plt.plot(x,y, color='#8B0000', label='função original f(x)')   #Ploting of graph of f
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np               # Import packages numpy ans matplotlib
import matplotlib.pyplot as plt

def g(x):    # Definition of function multisentence f(x+1)
    return f(x+1)
x=np.linspace(-3,1)    # Comprehenssion list for values of x and y
y=[g(i) for i in x]
print(y)

plt.plot(x,y,color='#008000', label='função transladada f(x+1)')   #Ploting of graph of f
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np               # Import packages numpy ans matplotlib
import matplotlib.pyplot as plt

def f(x):    # Definition of function multisentence f
    if (x>1):
        return 1
    elif (x<=1.0 and x>=-1.0):
        return x
    else:
        return -1 
x1=np.linspace(-2,2)    # Comprehenssion list for values of x and y
y1=[f(i) for i in x1]
#print(y) 

def g(x):    # Definition of function multisentence f(x+1)
    return f(x+1)
x2=np.linspace(-3,1)    # Comprehenssion list for values of x and y
y2=[g(i) for i in x2]
#print(y)
    

def h(x):    # Definition of function multisentence f(x+1)
    return f(x+2)
x3=np.linspace(-4,0)    # Comprehenssion list for values of x and y
y3=[h(i) for i in x3]
#print(y)
plt.figure(figsize=(8, 8))
plt.plot(x1,y1, color='#8B0000', label='função original f(x)')   #Ploting of graph of f
plt.plot(x2,y2,color='#008000', label='função transladada f(x+1)')   #Ploting of graph of f(x+1)
plt.plot(x3,y3,color='#FFD700', label='função transladada f(x+2)')   #Ploting of graph of f(x+2)
plt.grid()
plt.legend()
plt.show()
```
