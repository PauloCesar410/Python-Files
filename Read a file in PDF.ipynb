```python
import PyPDF2
import re
```

```python
pdf_file  = open('CÓDIGO PENAL ART.1-120.pdf','rb')  #Abre o arquivo
```

```python
read_pdf = PyPDF2.PdfFileReader(pdf_file)  #Faz a leitura do arquivo usando a biblioteca
```

```python
number_of_pages=read_pdf.getNumPages #número de páginas
```

```python
page=read_pdf.getPage(0-29) #lê a primeira página
```

```python
page_content = page.extractText()  #extrai apenas o texto
```

```python
parsed = ''.join(page_content) #faz a junção das linhas
```

```python
print(parsed)  #Não elimina as quebras
```

```python
parsed =re.sub('n','',parsed) #remove a quebra de linhas
```

```python
print("Após eliminar as quebras")
```

```python
print(parsed)
```
