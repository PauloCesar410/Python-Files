```python
coeficientes=(-0.0625, 0.0, 0.5625, 1.0, 0.5625, 0.0, -0.0625) # Interpolet de gênero 4
```

```python
def interpolet_escala(x):
    interpolet_val=0
    if ((x==-3) or (x==3) or (x==2) or (x==-2) or (x==1) or (x==-1)):
        return 0
    elif (x<-3 or x>3):
        return 0
    elif (x==0):
        return 1
    elif ((x>-3 and x<-2) or (x>-2 and x<-1) or (x>-1 and x<0) or (x>0 and x<1) or (x>1 and x<2) or (x>2 and x<3)):
        for m in range(0,7,1):
             interpolet_val += coeficientes[m]*interpolet_escala(2*x-m+3)
        return interpolet_val
```

```python
print(interpolet_escala(0.5))
```

```python
import matplotlib.pyplot as plt
x=[-3.0,-2.5,-2.0,-1.5,-1.0,-0.75,-0.5,0.0,0.5,0.75,1.0,1.5,2.0,2.5,3.0]    # Comprehenssion list for values of x and y
y=[interpolet_escala(i) for i in x]
#print(y)

plt.plot(x,y,color='#8B0000', label='função interpolet phi(x)')   #Ploting of graph of wavelet interpolet
plt.grid()
plt.legend()
plt.show()
```
