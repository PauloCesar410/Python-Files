```python
import numpy as np
def f(x):
    return np.sin(x)
```

```python
valor = int(input("Digite o valor do ângulo"))
angulo = (np.pi*valor/180)
print(np.sin(angulo))
```

```python
import matplotlib.pyplot as plt
x=np.linspace(-15,15)    # Comprehenssion list for values of x and y
y=[f(i) for i in x]
plt.figure(figsize=(5, 5))
plt.title ('Gráfico da função seno')
plt.plot(x,y, color='#8B0000', label='função seno')   #Ploting of graph of f
plt.grid()
plt.legend()
plt.show()
```
