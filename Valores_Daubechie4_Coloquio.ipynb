```python
#Estudos com Reuber
f=0.915*10**(9)
const = 0.58*10**(-10)
epsilonw_infinito=4.9
epsilonw0=80.10
epsilonlinha_fw = epsilonw_infinito + ((epsilonw0-epsilonw_infinito)/(1+(const*f)**2))
print(epsilonlinha_fw)
```

```python
#Estudos com Reuber
alpha= 0.65
rhos=2.62
rhob=1.61
mv=0.01
epsilions=(1+0.44*rhos)**2 -0.062
epsilonlinha_m = (1+ (rhob/rhos)*((epsilions)**(alpha)-1)+(mv)**(betalinha1)*(epsilonlinha_fw)**(alpha)-mv)**(1/alpha)
print(epsilonlinha_m)
```

```python
#Estudos com Reuber
def epsilonlinha_m(mv):
    return 1.15*( 1+ (rhob/rhos)*((epsilions)**(alpha)-1)+(mv)**(betalinha1)*(epsilonlinha_fw)**(alpha)-mv)**(1/alpha)-0.68
```

```python
#Estudos com Reuber
sigmaeff= 0.0467+0.2204*rhob-0.4111*linear1+0.6614*linear2
e0= 8.854*10**(-12)
epsilonduaslinhas_fw= (f*const*(epsilonw0-epsilonw_infinito)/(1+(f*const)**2))+(sigmaeff*(rhos-rhob)/(2*3.14*f*e0)*(rhos*mv))
```

```python
def epsilonduaslinhas_m(mv):
    return ((mv)**(betalinha2)*(epsilonduaslinhas_fw)**(alpha))**(1/alpha)
```
